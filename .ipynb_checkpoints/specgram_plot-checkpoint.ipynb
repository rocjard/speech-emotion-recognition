```python
import matplotlib.pyplot as plt
from scipy.io import wavfile
import numpy as np

samplingFrequency, signalData = wavfile.read('engenharia.wav')

signal = signalData
N = signalData.size

SAMPLE_RATE = 44100  # Hertz
DURATION = N/samplingFrequency  # Seconds

x = np.arange(N)/samplingFrequency

plt.plot(x, signalData)
```

```python
from scipy.fft import rfft, rfftfreq

yf = rfft(signal)
xf = rfftfreq(N, 1 / SAMPLE_RATE)

plt.plot(xf, np.abs(yf))
#plt.xlim([0,2000])
plt.show()
```

```python
import librosa

y, sr = librosa.load('engenharia.wav')
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True)

D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
img = librosa.display.specshow(D, y_axis='linear', x_axis='time',
                               sr=sr, ax=ax[0])
ax[0].set(title='Linear-frequency power spectrogram')
ax[0].label_outer()

hop_length = 1024
D = librosa.amplitude_to_db(np.abs(librosa.stft(y, hop_length=hop_length, win_length=1024)),
                            ref=np.max)
librosa.display.specshow(D, y_axis='log', sr=sr, hop_length=hop_length,
                         x_axis='time', ax=ax[1])
ax[1].set(title='Log-frequency power spectrogram')
ax[1].label_outer()
fig.colorbar(img, ax=ax, format="%+2.f dB")
```

```python
y, sr = librosa.load('engenharia.wav')
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True)

samplingFrequency, signalData = wavfile.read('engenharia.wav')

signal = signalData
N = signalData.size

SAMPLE_RATE = 44100  # Hertz
DURATION = N/samplingFrequency  # Seconds

x = np.arange(N)/samplingFrequency

ax[0].plot(x, signalData/20000)


#ax[0].set(title='Linear-frequency power spectrogram')
ax[0].set(ylabel="Amplitude")
ax[0].label_outer()
ax[0].set_ylim([-1,1])
hop_length = 56
#D = librosa.amplitude_to_db(np.abs(librosa.stft(y, hop_length=hop_length)),
#                            ref=np.max)
#librosa.display.specshow(D, y_axis='log', sr=sr, hop_length=hop_length,
#                         x_axis='time', ax=ax[1])
#ax[1].set(title='Log-frequency power spectrogram')
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
img = librosa.display.specshow(D, y_axis='linear', x_axis='time',
                               sr=sr, ax=ax[1])

ax[1].set(xlabel="Tempo (s)")
ax[1].set(ylabel="Frequnica (Hz)")
ax[1].set_ylim([0, 11000])
#ax[1].set_xlim([0, 12.9])
ax[1].label_outer()
fig.colorbar(img, ax=ax[0], format="%+2.f dB")
fig.colorbar(img, ax=ax[1], format="%+2.f dB")
```

```python
y, sr = librosa.load('engenharia.wav')

S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128,
                                    fmax=8000)

fig, ax = plt.subplots()
S_dB = librosa.power_to_db(S, ref=np.max)
img = librosa.display.specshow(S_dB, x_axis='time',
                         y_axis='mel', sr=sr,
                         fmax=8000, ax=ax)
fig.colorbar(img, ax=ax, format='%+2.0f dB')
#ax.set(title='Mel-frequency spectrogram')
ax.set(xlabel="Tempo (s)")
ax.set(ylabel="Frequncia (mel)")
```
